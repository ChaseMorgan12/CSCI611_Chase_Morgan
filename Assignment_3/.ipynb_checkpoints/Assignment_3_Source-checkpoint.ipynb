```python
from ultralytics import YOLO
import matplotlib.pyplot as plt

model = YOLO("yolov8n.pt")

#get results from the pre-trained model and save them to disk.
model.predict(
    source="dataset/test/images",
    save=True,
    conf=0.5,
   imgsz=640
)

#train model to hopefully get better results.
model.train(
    data="dataset/data.yaml",
    epochs=20,
    imgsz=640,
    batch=16,
    cache=False
)

model.predict(
    source="dataset/test/images",
    save=True,
    conf=0.5,
    imgsz=640
)
```
